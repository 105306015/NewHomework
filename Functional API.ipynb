# 讀入套件

```python
import matplotlib.pyplot as plt
import matplotlib.image as pimg
import numpy as np

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Flatten


from tensorflow.keras.layers import Conv2D, MaxPool2D, GlobalAveragePooling2D

from tensorflow.keras.optimizers import SGD, Adam


from tensorflow.keras import datasets
from tensorflow.keras.utils import to_categorical

from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input

from tensorflow.keras.layers import concatenate, add
```

```python
# Load Fashine MNIST
(x_train, y_train0), (x_test, y_test0) = datasets.fashion_mnist.load_data()

# Reshape size
x_train = x_train.reshape(-1, 28*28)
x_test = x_test.reshape(-1, 28*28)

# Normalize the range of featurs
x_train = x_train / x_train.max()
x_test = x_test / x_test.max()

# One-hot encoding
y_train = to_categorical(y_train0, 10)
y_test = to_categorical(y_test0, 10)
```

```python
f_1 = Dense(500, activation='sigmoid')
f_2 = Dense(300, activation='sigmoid')
f_4 = Dense(400, activation='relu')
f_3 = Dense(500, activation='relu')
f_5 = Dense(10, activation='softmax')
```

```python
x = Input(shape=(784,))
```

```python
h_1 = f_1(x)
h_2 = f_2(h_1)
z = f_4(h_1)
u = concatenate([h_2, z])
h_3 = f_3(u)
y = f_5(h_3)
```

```python
model = Model(x, y)
model.summary()
```

```python
model.compile(loss='mse', 
              optimizer=Adam(), 
              metrics=['accuracy'])
```

batch_size改為150，epochs改為5

```python
model.fit(x_train, y_train, 
          batch_size=150, 
          epochs=5)
```

```python
score = model.evaluate(x_train, y_train, batch_size=10000)
print(f'Loss: {score[0]}')
print(f'準確率: {score[1]*100}')
```
